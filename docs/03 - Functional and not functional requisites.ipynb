# Requisitos del Proyecto VerdeNica

## Requisitos Funcionales

1. **Registro de Usuarios**
   - Los usuarios deben poder registrarse en la plataforma proporcionando información básica (nombre, email, contraseña).
  
2. **Autenticación**
   - Los usuarios deben poder iniciar sesión y recuperar sus credenciales en caso de olvidarlas.

3. **Publicaciones Ambientales**
   - Los usuarios deben poder crear, editar y eliminar publicaciones relacionadas con la flora, fauna y medio ambiente.
   - Las publicaciones pueden recibir **like** o **dislike** por parte de otros usuarios.

4. **Denuncias Ambientales**
   - Los usuarios deben poder denunciar violaciones ambientales de forma anónima, creando un registro que se almacenará en la base de datos.

5. **Acceso a Información Legal**
   - La plataforma debe proporcionar acceso a documentos relevantes, como el Acuerdo de París y leyes ambientales de Nicaragua en formato PDF.

6. **Campañas**
   - Los usuarios deben poder crear, visualizar y participar en campañas ambientales.
   - Se deben permitir votaciones (like/dislike) en las campañas existentes.

7. **Foro de Discusión**
   - La plataforma debe incluir un foro donde los usuarios puedan iniciar discusiones organizadas en categorías (preguntas, debates, hilos de conversación).

8. **Biodiversidad en Nicaragua**
   - Se debe incluir una sección dedicada a la flora y fauna del país con imágenes, nombres y descripciones breves.

9. **Reconocimiento de Especies con IA**
   - Los usuarios deben poder subir imágenes y recibir información sobre especies de animales o plantas utilizando IA.

10. **Sistema de Suscripción**
    - Los usuarios deben poder suscribirse a notificaciones ingresando su correo electrónico.

## Requisitos No Funcionales

1. **Usabilidad**
   - La interfaz debe ser intuitiva y fácil de navegar para usuarios de diferentes niveles de habilidad tecnológica.

2. **Accesibilidad**
   - La plataforma debe cumplir con los estándares de accesibilidad, permitiendo que usuarios con discapacidades puedan interactuar sin dificultades.

3. **Seguridad**
   - La información del usuario, incluidas las credenciales y denuncias, debe ser almacenada de forma segura para garantizar la privacidad.

4. **Rendimiento**
   - La aplicación debe ser capaz de manejar múltiples usuarios simultáneamente sin degradar el rendimiento.

5. **Compatibilidad**
   - La plataforma debe ser accesible desde diferentes dispositivos (móviles, tablets, computadoras) y navegadores web.

6. **Mantenibilidad**
   - El código y la arquitectura de la aplicación deben ser mantenibles y fáciles de actualizar.

7. **Escalabilidad**
   - La aplicación debe poder escalar para manejar un aumento en el número de usuarios y publicaciones.

8. **Disponibilidad**
   - La plataforma debe estar disponible 24/7, garantizando el acceso continuo a los usuarios.

9. **Documentación**
   - Debe existir documentación clara y accesible sobre el uso de la plataforma y su mantenimiento técnico.
