# Manual de Uso de VerdeNica

## Descripción
VerdeNica es un proyecto innovador diseñado para promover buenas prácticas medioambientales con un enfoque comunitario. Permite a los nicaragüenses participar activamente en la creación de campañas ambientales, documentar cualquier dato relacionado con la flora, fauna o medio ambiente en general a través de publicaciones, y denunciar violaciones ambientales de forma anónima, garantizando la integridad del usuario y el contenido que se publica en la plataforma.

## Requisitos Previos
Para utilizar la plataforma solo necesitas un teléfono móvil, tablet o computadora con conexión a internet. Accede a la URL del sitio en tu navegador web favorito o mediante un sistema de búsqueda como Google o Edge.

## Principales Interfaces y sus Funciones
Cada pantalla en la plataforma incluye un menú de tres barras ubicado en la parte superior izquierda. A través de este menú puedes acceder a las principales páginas del sitio, como **Inicio**, **Foro**, **Campañas**, **Empresas Sostenibles**, **Ayuda**, y más.

### Inicio
- La pantalla principal muestra un encabezado verde con el slogan de la plataforma.
- Un menú horizontal celeste contiene cinco elementos: **Empresas Sostenibles**, **Campañas**, **Foro**, **Ayuda**.
- La sección **main** incluye:
  - Botones para acceder al **Acuerdo de París** y al compendio de leyes de Nicaragua en formato PDF.
  - Botón de **Denuncias** que lleva a la página de denuncias existentes y para crear nuevas denuncias.

![Captura de Pantalla - Página de Inicio](../static/images/inicio-uno.png)    

  - Botones para **Crear Nueva Publicación**, **Biodiversidad en Nicaragua**, y **Reconocimiento de Especies con IA**.
  - Listas de causas y consecuencias del cambio climático.
  - Información sobre **Mitigación** y **Adaptación**.
  - Sección sobre **Sostenibilidad** con un enlace a una página dedicada.
  - **Acciones de Mitigación** que son enlaces a detalles sobre acciones aplicables en la vida diaria.

### Cómo Funciona la Suscripción por Correo
El usuario ingresa su email y hace clic en **Suscribirse**. Este formulario utiliza **Formspree** para enviar un correo notificando que un usuario ha enviado el formulario junto con su dirección de correo electrónico.
